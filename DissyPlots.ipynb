# **Setup**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('out.csv')
```

## **View data**

# **Functions**

```python
# Function to convert TIME to milliseconds
def time_to_milliseconds(time_str):
    minutes, seconds = time_str.split(':')
    seconds, milliseconds = seconds.split('.')
    total_milliseconds = (int(minutes) * 60 + int(seconds)) * 1000 + int(milliseconds)
    return total_milliseconds
```

# **Plots**

```python
# Convert TIME to milliseconds and add as a new column
df['TimeInMilliseconds'] = df['TIME'].apply(time_to_milliseconds)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['TimeInMilliseconds'], df['%CPU'], marker='o', linestyle='-', color='b')
plt.title('Time (ms) vs CPU Consumption')
plt.xlabel('Time (milliseconds)')
plt.ylabel('%CPU Consumption')
plt.grid(True)
plt.show()
```
